```python
import numpy as np
import pandas as pd

data = pd.read_csv('plant_data_export.csv')
data = data.set_index('Unnamed: 0')
data.head(5)
```

```python
data.isnull().sum()
```

```python
#first phase of data cleaning
#first encode the categorical values to numerical values
#Since (data.Microgreen, data.Flowering, data.Bulbing, data.Stratification) have just two categories in their input data i.e. True or False
#I use LabelEncoder to convert the data to a binary state of 1 and 0 for True and False respectively.

from sklearn.preprocessing import LabelEncoder

l_encoder = LabelEncoder()
data['Microgreen'] = l_encoder.fit_transform(data['Microgreen'])
data['Flowering'] = l_encoder.fit_transform(data['Flowering'])
data['Bulbing'] = l_encoder.fit_transform(data['Bulbing'])
data['Stratification'] = l_encoder.fit_transform(data['Stratification'])
#data.tail(10)
```

```python
#For the NaN cells, I would assume it means zero presence of vitamins in these plants, so I fill with 0
data['Vitamin A'] = data['Vitamin A'].fillna('0')
data['Vitamin D'] = data['Vitamin D'].fillna('0')
data['Vitamin E'] = data['Vitamin E'].fillna('0')
data['Vitamin K'] = data['Vitamin K'].fillna('0')
data['Vitamin C'] = data['Vitamin C'].fillna('0')
data['Vitamin B1'] = data['Vitamin B1'].fillna('0')

data.head()
```

```python
#For the different vitamins, available values include High, Medium and Low
#I convert using One-Hot encoding because they are nominal and contain more than two classes of labels
#Also another reason is OneHotEncoding is used because there is really no relationship between the values in each vitamin feature
#and I don't want the model learning/or creating one during training.
#For the NaN cells, I would assume it means zero presence of vitamins in these plants, so I fill with 0
#data.fillna(0)
#data.head()
from sklearn.preprocessing import OneHotEncoder

h_encoder = OneHotEncoder()

#vitamin_list = ['Vitamin A', 'Vitamin D']#, 'Vitamin E', 'Vitamin K', 'Vitamin C', 'Vitamin B1']

#for i in vitamin_list:
#data[i] = h_encoder.fit(data[i].reshape(-1,1))

data['Vitamin A'] = l_encoder.fit_transform(data['Vitamin A'])
#data['Vitamin A'] = h_encoder.fit_transform(data[['Vitamin A']])

data['Vitamin D'] = l_encoder.fit_transform(data['Vitamin D'])
#data['Vitamin D'] = h_encoder.fit_transform(data[['Vitamin D']])

data['Vitamin E'] = l_encoder.fit_transform(data['Vitamin E'])
#data['Vitamin E'] = h_encoder.fit_transform(data[['Vitamin E']])

data['Vitamin K'] = l_encoder.fit_transform(data['Vitamin K'])
#data['Vitamin K'] = h_encoder.fit_transform(data[['Vitamin K']])

data['Vitamin C'] = l_encoder.fit_transform(data['Vitamin C'])
#data['Vitamin C'] = h_encoder.fit_transform(data[['Vitamin C']])

data['Vitamin B1'] = l_encoder.fit_transform(data['Vitamin B1'])
#data['Vitamin B1'] = h_encoder.fit_transform(data[['Vitamin B1']])
```

```python
#Before feeding the model, i normalize the data in a bid to avoid the model from being biased towrds certain feature

from sklearn.preprocessing import normalize
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
#normalizer = normalize()

data_scaled = scaler.fit_transform(data)
#data_normed = normalizer.fit_transform(data)
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10, 7))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(data, method='average'))
plt.axhline(y=65, color='r', linestyle='--')
```

```python
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='average')  
ac_fitting = ac.fit_predict(data_scaled)
ac.labels_
```
