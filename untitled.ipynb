```python
#import packages and read the data into pandas
import numpy as np
import pandas as pd

data = pd.read_csv('plant_data_export.csv')
data = data.set_index('Unnamed: 0')
data.head(5)
```

```python
#first phase of data cleaning
#first encode the categorical values to numerical values
#Since (data.Microgreen, data.Flowering, data.Bulbing, data.Stratification) have just two categories in their input data i.e. True or False
#I use LabelEncoder to convert the data to a binary state of 1 and 0 for True and False respectively.

from sklearn.preprocessing import LabelEncoder

l_encoder = LabelEncoder()
data['Microgreen'] = l_encoder.fit_transform(data['Microgreen'])
data['Flowering'] = l_encoder.fit_transform(data['Flowering'])
data['Bulbing'] = l_encoder.fit_transform(data['Bulbing'])
data['Stratification'] = l_encoder.fit_transform(data['Stratification'])
#data.tail(10)
```

```python
#For the NaN cells, I would assume it means zero presence of vitamins in these plants, so I fill with string 0
data['Vitamin A'] = data['Vitamin A'].fillna('0')
data['Vitamin D'] = data['Vitamin D'].fillna('0')
data['Vitamin E'] = data['Vitamin E'].fillna('0')
data['Vitamin K'] = data['Vitamin K'].fillna('0')
data['Vitamin C'] = data['Vitamin C'].fillna('0')
data['Vitamin B1'] = data['Vitamin B1'].fillna('0')

data.head()
```

```python
#For the different vitamins, available values include High, Medium and Low
#I convert using label encoding

data['Vitamin A'] = l_encoder.fit_transform(data['Vitamin A'])
data['Vitamin D'] = l_encoder.fit_transform(data['Vitamin D'])
data['Vitamin E'] = l_encoder.fit_transform(data['Vitamin E'])
data['Vitamin K'] = l_encoder.fit_transform(data['Vitamin K'])
data['Vitamin C'] = l_encoder.fit_transform(data['Vitamin C'])
data['Vitamin B1'] = l_encoder.fit_transform(data['Vitamin B1'])
```

```python
#to display the importance of each feature in relation to others 

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

X = data.iloc[:,5:]  #independent columns
y = data.iloc[:,0]    #target column i.e plant id #apply SelectKBest class to extract top 10 best features

bestfeatures = SelectKBest(score_func=chi2, k=10)

fit = bestfeatures.fit(X,y)

dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)

#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  #naming the dataframe columns
print(featureScores.nlargest(10,'Score'))  #print 10 best features
```

```python
#visualizing importance of the features

from sklearn.ensemble import ExtraTreesClassifier

model = ExtraTreesClassifier()
model.fit(X,y)
print(model.feature_importances_) #use inbuilt class feature_importances of tree based classifiers

#plot graph of feature importances for better visualization
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.show()
```

```python
from sklearn.preprocessing import normalize

data_scaled = normalize(data)
data_scaled = pd.DataFrame(data_scaled, columns=data.columns)
#data_scaled.head()
```

```python
#using dendogram to visualize how the points are close to each other
#while the vertical heights showcases the closeness between clusters
#setting a threshold of 65, it cuts across clusters with maximum distance, thereby obtaining the n_clusters to be 5.
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10, 7))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(data_scaled, method='ward'))
plt.axhline(y=2.5, color='r', linestyle='--')
```

```python
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')  
ac.fit_predict(data_scaled)
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(data_scaled['Plant ID'], data_scaled['Vitamin K'], c=ac.labels_)
plt.xlabel('Plant ID')
plt.ylabel('Vitamin K')
plt.title('Agglomerative Clustering for Prioritizing Plant Support')
```
